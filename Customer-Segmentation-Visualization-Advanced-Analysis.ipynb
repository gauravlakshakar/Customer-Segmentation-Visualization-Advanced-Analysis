# Tasks 4: Customer Segmentation Visualization


```python
#tenure categories
df['tenure_group'] = pd.cut(df['tenure'], bins=[0, 12, 36, df['tenure'].max()], labels=['0-12 months', '13-36 months', '37+ months'])
df
```

```python
df['tenure_group'].value_counts()
```

```python
# Donut Chart
import plotly.express as px
fig = px.pie(df, names='tenure_group', title='Customer Distribution by Tenure', hole=0.3) 
fig.show()
```

```python
# Group data by tenure and calculate average monthly charges
tenure_charges = df.groupby('tenure_group')['monthlycharges'].mean().reset_index()
```

```python
# Bar chart
import plotly.graph_objects as go
fig = go.Figure([go.Bar(x=tenure_charges['tenure_group'], y=tenure_charges['monthlycharges'], text=tenure_charges['monthlycharges'])])
```

```python
fig.update_layout(title='Avg Charges by Tenure', xaxis_title='Tenure', yaxis_title='Charges')
fig.show()
```
